```python
df_train = df[df.index<='2023-08-01']
df_test = df[df.index>'2023-08-01']
plt.plot(df_train,color='red')
plt.plot(df_test,color='green')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

seasonal = seasonal_decompose(df_train,period=7)

trend = seasonal.trend
estacionalidad = seasonal.seasonal
error = seasonal.resid

plt.plot(trend)
plt.plot(estacionalidad)
plt.plot(error,color='purple')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

np.random.seed(42)
df = np.random.normal(0, 1, 100).cumsum()

def test_adfuller(series):
    result = adfuller(series)
    print('--- Resultado del Test ADF ---')
    print(f'Estadístico ADF: {result[0]:.4f}')
    print(f'p-valor: {result[1]:.4f}')
    print(f'Retrasos usados (Lags): {result[2]}')
    print(f'Número de observaciones: {result[3]}')
    print('Valores Críticos:')
    for key, value in result[4].items():
        print(f'   {key}: {value:.4f}')

    if result[1] <= 0.05:
        print("\nConclusión: Rechazamos H0. La serie es ESTACIONARIA.")
    else:
        print("\nConclusión: No rechazamos H0. La serie tiene raíz unitaria (es NO ESTACIONARIA).")

test_adfuller(df)
```

1) En el caso de esta serie temporal hablamos de un tensor de 2 dimensiones.
2) Se observa una tendencia lineal al alza en el periodo de estudio.
3) Se realizo test de raiz unitaria ADF y la serie tiene raiz unitaria por lo que se concluye que es no estacionaria.
4) No se observa estacionalidad en la serie de tiempo.
5) No se observa presencia de ruido en la serie de tiempo.

```python
from pmdarima import auto_arima

model = auto_arima(df_train,seasonal=False,trace=True)
```

```python
sales_pred = model.predict(34)
sales_pred
```

```python
plt.plot(df_test,color='red')
plt.plot(sales_pred,color='green')
plt.show()
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score

error_2 = root_mean_squared_error(df_test,sales_pred)
r2 = r2_score(df_test,sales_pred)

print(f'El error cuadratico medio en raiz es de {error_2} y el R2 es de {r2}')
```


Conclusión: Basado en los parametros del modelo se puede aseverar que es un 98% predictivo de las ventas.
